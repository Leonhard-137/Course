# 线性判别函数的几何意义（精要版）

## 一、线性判别函数定义

$$
g(x) = w^T x + w_0
$$

分类规则：

$$
\begin{cases}
g(x) > 0 \Rightarrow x \in \omega_1 \\
g(x) \le 0 \Rightarrow x \in \omega_2
\end{cases}
$$

---

## 二、几何意义

- \( g(x) = 0 \)：表示一个**超平面**（二维为直线，三维为平面）。  
- \( w \)：为平面的**法向量**，决定方向。  
- \( w_0 \)：为平面的**截距/偏置项**，决定平面距离原点的远近。  
- \( g(x) \)：表示样本点到平面的**带符号距离 × 法向量长度**。

---

## 三、点到平面的距离推导

平面方程：
$$
w^T x + w_0 = 0
$$

平面上点 \( x_p \) 满足：
$$
w^T x_p + w_0 = 0
$$

任意点 \( x \)：
$$
x = x_p + (x - x_p)
$$

对两边取内积得：
$$
w^T x = w^T x_p + w^T (x - x_p)
$$

因为 \( w^T x_p + w_0 = 0 \)，可得：

$$
g(x) = w^T x + w_0 = w^T (x - x_p)
$$

沿 \( w \) 方向的几何距离为：

$$
z = \frac{g(x)}{\|w\|}
$$

因此：

$$
|z| = \frac{|g(x)|}{\|w\|}
$$

---

## 四、偏置项与原点距离关系

超平面到原点的距离：

$$
|d| = \frac{|w_0|}{\|w\|}
$$

---

## 五、核心关系式总结

$$
g(x) = \|w\| \cdot z
$$

或等价地：

$$
|z| = \frac{|g(x)|}{\|w\|}, \quad |d| = \frac{|w_0|}{\|w\|}
$$

---

## 六、几何结论

- \( g(x)=0 \)：分类边界（超平面）  
- \( g(x)>0 \)：点在法向量 \( w \) 方向一侧  
- \( g(x)<0 \)：点在法向量反方向  
- \( |g(x)| \)：点到平面的距离 × 法向量长度
